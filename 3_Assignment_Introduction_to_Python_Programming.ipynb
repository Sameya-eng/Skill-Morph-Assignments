## Assignment 1: Basic If-Else Statement

**Task:** Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.

```python
age=int(input("enter your age"))
if age>=18:
    print("eligible to vote")
else:
    print("not eligible")
```

## Assignment 2: Multiple Conditions

**Task:** Write a Python program that asks for the user's age and prints the appropriate message:
- If the user is 18 or older, print 'Adult'
- If the user is 13 or older but under 18, print 'Teenager'
- If the user is younger than 13, print 'Child'

```python
age=int(input("enter your age"))
if age>=18:
    print("Adult")
elif age>=13 and age<18:
    print("Teenager")
else:
    print("Child")
```

## Assignment 4: Conditional Expressions

**Task:** Use a ternary operator to determine if a number entered by the user is even or odd. The program should print 'Even Number' or 'Odd Number' based on the input.

```python
num = int(input("Enter a number: "))

result = "Even Number" if num % 2 == 0 else "Odd Number"

print(result)
```

# Python Programming Assignments - Loops and Functions

## Assignment 1: Sum of Natural Numbers

**Task:** Write a Python function `sum_of_natural_numbers(n)` that returns the sum of all natural numbers up to `n`. Use a loop inside the function.

```python
def sum_of_natural_numbers(n):
    sum=0
    for i in range(1,n+1):
        sum+=i
    return sum
    print(sum_of_natural_numbers(5))
```

## Assignment 2: Factorial of a Number

**Task:** Write a Python function `factorial(n)` that calculates and returns the factorial of `n`. Use a loop to calculate the factorial.

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
num = int(input("Enter a number: "))
result = factorial(num)
print("Factorial of", num, "is:", result)
```

## Assignment 3: Multiplication Table

**Task:** Write a Python function `multiplication_table(n)` that prints the multiplication table of `n` up to 10.

```python
def multiplication_table(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

num = int(input("Enter a number: "))
multiplication_table(num)
```
